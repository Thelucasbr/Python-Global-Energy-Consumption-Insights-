```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carrega o CSV (caso ainda n√£o tenha feito)
df = pd.read_csv('/content/World Energy Consumption.csv')
df_countries = df[df['iso_code'].notna()]
latest = df_countries.sort_values('year').groupby('country').tail(1)

# Fun√ß√£o para extrair top 5 por tipo de energia
def get_top5(df, column):
    return df[['country', 'year', column]].dropna().sort_values(by=column, ascending=False).head(5).reset_index(drop=True)

# Top 5 para cada tipo de energia
top5_nuclear = get_top5(latest, 'nuclear_consumption')
top5_solar = get_top5(latest, 'solar_consumption')
top5_wind = get_top5(latest, 'wind_consumption')
top5_hydro = get_top5(latest, 'hydro_consumption')
top5_fossil = get_top5(latest, 'fossil_fuel_consumption')

# Fun√ß√£o para criar gr√°fico
def plot_energy_bar(data, column, title, color):
    plt.figure(figsize=(10, 6))
    sns.barplot(data=data, x=column, y="country", palette=[color]*5)
    plt.title(title, fontsize=18, fontweight='bold')
    plt.xlabel("Consumption (TWh)", fontsize=12)
    plt.ylabel("Country", fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(axis='x', linestyle='--', alpha=0.6)
    sns.despine(left=True, bottom=True)
    plt.tight_layout()
    plt.show()

# üî¥ Nuclear
plot_energy_bar(top5_nuclear, 'nuclear_consumption', 'Top 5 Countries by Nuclear Energy Consumption (2022)', '#D62728')

# ‚òÄÔ∏è Solar
plot_energy_bar(top5_solar, 'solar_consumption', 'Top 5 Countries by Solar Energy Consumption (2022)', '#FDB813')

# üå¨Ô∏è Wind
plot_energy_bar(top5_wind, 'wind_consumption', 'Top 5 Countries by Wind Energy Consumption (2022)', '#76C7C0')

# üíß Hydro
plot_energy_bar(top5_hydro, 'hydro_consumption', 'Top 5 Countries by Hydroelectric Energy Consumption (2022)', '#1F77B4')

# üõ¢Ô∏è Fossil
plot_energy_bar(top5_fossil, 'fossil_fuel_consumption', 'Top 5 Countries by Fossil Fuel Energy Consumption (2022)', '#C44E52')
```
